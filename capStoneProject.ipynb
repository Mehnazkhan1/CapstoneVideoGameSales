# Regression analysis

```python
import statsmodels.api as sm # for statistical tests
```

```python
regression_df = df[['NA_Sales','EU_Sales','JP_Sales','Other_Sales','Global_Sales','Critic_Score','Critic_Count','User_Count']] # data for use of regression tes
```

```python
dvar = regression_df['Global_Sales']  # dependand variable
ivar = regression_df.drop(columns='Global_Sales') # dropping dependand variable
ivar['constant'] = np.ones((ivar.shape[0],)) #add a constant column for the regression, more on this later
```

```python
lin_reg = sm.OLS(dvar, ivar)
reg_results = lin_reg.fit()
print(reg_results.summary())
```

```python
regression_df2 = df[['EU_Sales','JP_Sales','Other_Sales','Global_Sales','Critic_Score','Critic_Count']]
```

```python
dvar2 = regression_df2['Global_Sales'] # its our dependant variable
ivar2 = regression_df2.drop(columns='Global_Sales') #dropping dependant variable
ivar2['constant'] = np.ones((ivar2.shape[0],)) #add a constant column for the regression, more on this later
```

```python
lin_reg = sm.OLS(dvar2, ivar2)
reg_results = lin_reg.fit()
print(reg_results.summary())
```
